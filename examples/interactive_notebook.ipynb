## Installation

First, install MAVERIC from the GitHub repository:

```python
# Import required libraries
from maveric import MAVERIC, MAVERICConfig
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Initialize MAVERIC

```python
# Configure MAVERIC
config = MAVERICConfig(
    cache_base_dir="./maveric_cache",
    clip_model="ViT-B/32"
)

maveric = MAVERIC(config)
```

## 2. Load or Retrieve Data

```python
# Option 1: Load existing results
# data = pd.read_json("path/to/retrieval_results.json")

# Option 2: Retrieve new data
retrieval_result = maveric.retrieve(
    dataset_name="react-vl/react-retrieval-datasets",
    target_dataset="cifar10",
    num_samples=10000  # Small sample for demo
)
```

## 3. Launch Interactive Dashboard

```python
# Launch the interactive quality control dashboard
dashboard = maveric.launch_dashboard(retrieval_result)
dashboard
```

## 4. Visualize Results

```python
# After adjusting thresholds in the dashboard, visualize the results
figures = maveric.visualize_retrieval(
    retrieval_result,
    metrics=['sharpness_score', 'consistency', 'color_score']
)

# Display figures
for metric, fig in figures.items():
    plt.show()
```

## 5. Export Configuration

```python
# After finding optimal thresholds, export the configuration
optimal_thresholds = {
    'weighted_class_score': 0.5,
    'consistency': 0.8,
    'sharpness_score': 0.85,
    'resolution_score': 0.4,
    'color_score': 0.7
}

# Apply the optimized thresholds
quality_result = maveric.quality_control(
    retrieval_result,
    thresholds=optimal_thresholds,
    balance_strategy='median'
)

print(f"Final dataset: {quality_result.filtered_count} samples")
print(f"Retention rate: {quality_result.retention_rate:.1%}")
```
